```python
import tensorflow as tf
a = tf.constant([3,6])
b = tf.constant([2,2])
with tf.Session() as sess:
    print ("add:")
    print (sess.run(tf.add(a,b)))
    print ("add_n:")
    print (sess.run(tf.add_n([a,b,b])))
    print ("mul:")
    print (sess.run(tf.mul(a,b)))
    print ("matmul:")
    #print (sess.run(matmul(a,b)))
    print ("reshape:")
    print (sess.run(tf.reshape(a,[1,2])))
    print ("matmul:")
    print (sess.run(tf.matmul(tf.reshape(a,[1,2]),tf.reshape(b,[2,1]))))
    print ("div:")
    print (sess.run(tf.div(a,b)))
    print ("mod:")
    print (sess.run(tf.mod(a,b)))
```

```python
#给占位符赋值
#tf.placeholder(dtype, shape=None, name=None)
# create a placeholder of type float 32-bit, shape is a vector of 3 elements
a = tf.placeholder(tf.float32, shape=[3])
# create a constant of type float 32-bit, shape is a vector of 3 elements
b = tf.constant([5, 5, 5], tf.float32)
# use the placeholder as you would a constant or a variable
c = a + b # Short for tf.add(a, b)
with tf.Session() as sess:
# feed [1, 2, 3] to placeholder a via the dict {a: [1, 2, 3]}
# fetch value of c
    print (sess.run(c, {a: [1, 2, 3]})) # the tensor a is the key, not the string ‘a’
```

```python
#正常加载
x = tf.Variable(10, name='x')
y = tf.Variable(20, name='y')
z = tf.add(x, y) # you create the node for add node before executing the graph
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    writer = tf.summary.FileWriter('./my_graph/l2', sess.graph)
    for _ in range(10):
        sess.run(z)
    writer.close()
    

##Lazy loading:
x = tf.Variable(10, name='x')
y = tf.Variable(20, name='y')
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    writer = tf.summary.FileWriter('./my_graph/l2', sess.graph)
    for _ in range(10):
        sess.run(tf.add(x, y)) # someone decides to be clever to save one line of code
    writer.close()
```
